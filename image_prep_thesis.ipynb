```python
from PIL import Image
import matplotlib.pyplot as plt
format_path = './images/text_image_{}_example.jpg'
caption_dict = {
    'house' : '"A house surrounded by greenery."',
    'dog': '"A Shiba Inu lying down."',
    'shoe' : '"I bought a new pair!"'
}
fig, ax = plt.subplots(1, len(list(caption_dict.keys())), figsize = (16,5))
image_size = (233,155)
for i, image_name in enumerate(caption_dict):
    caption = caption_dict[image_name]
    image_path = format_path.format(image_name)
    image = Image.open(image_path)
    image = image.resize(image_size)
    ax[i].set_title(caption, fontsize = 20)
    ax[i].imshow(image)
    ax[i].axis('off')
plt.tight_layout()
plt.savefig('./images/caption_examples.pdf', transparent = True, bbox_inches='tight')
```
